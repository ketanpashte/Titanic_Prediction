```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
import tensorflow as tf
print(tf.__version__)
```

```python
train_dataset = pd.read_csv('/content/drive/MyDrive/train.csv')
test_dataset = pd.read_csv('/content/drive/MyDrive/test.csv')
```

```python
train_dataset.head()
```

```python
train_dataset = train_dataset[['Pclass', 'Sex', 'Age', 'Survived', 'SibSp', 'Parch', 'Fare']].copy()
test_dataset = test_dataset[['PassengerId', 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare']].copy()
train_dataset.head()
```

```python
train_features = train_dataset.dropna()
train_labels = np.array(train_features.pop('Survived'))
print(train_labels)
train_features.head()
```

```python
test_dataset.head()
```

```python
test_features = test_dataset.copy()
# keep passenger ids for later
test_pids = test_features.pop('PassengerId')
test_features.head()
```

```python
age_mean = np.concatenate((
    np.array(train_features['Age'].dropna()),
    np.array(test_features['Age'].dropna())
)).mean()
print(age_mean)
```

```python
test_features['Age'].isna().sum()
```

```python
test_features['Age'].fillna(age_mean, inplace=True)
test_features['Age'].isna().sum()
```

```python
test_features['Sex'].replace({ 'male': 0, 'female': 1 }, inplace=True)
train_features['Sex'].replace({ 'male': 0, 'female': 1 }, inplace=True)
```

```python
train_features.head()
```

```python
normalizer = tf.keras.layers.Normalization(axis=-1)
normalizer.adapt(train_features)
print(normalizer(train_features[:10]))
```

```python
model = tf.keras.Sequential([
    normalizer,
    tf.keras.layers.Dense(6, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid'),
])
```

```python
model.predict(test_features[:10])
```

```python
model.compile(loss=tf.keras.losses.binary_crossentropy, optimizer='adam', metrics=['binary_accuracy'])
```

```python
model.fit(train_features, train_labels,
          epochs=205, batch_size=64,
          use_multiprocessing=True, verbose=True)
```

```python
model.predict(test_features[:20])
```

```python
predictions = model.predict(test_features)
```

```python
survived = list(map(lambda p: int(p[0]>=0.5),predictions))
print((predictions > 0.5))
for s, p in zip(survived[:20], predictions[:20]):
    print(p, s)
```

```python
test_pids.head()
```

```python
result = pd.DataFrame({
    'PassengerId': test_pids.array,
    'Survived': survived
})

result.head()
```

```python
result.to_csv('/content/drive/MyDrive/gender_submission.csv', index=False)
```
